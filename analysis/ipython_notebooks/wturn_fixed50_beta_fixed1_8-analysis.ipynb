This notebook looks at the fits for the run where we fixed $\lambda_{turn}=50$ $\mu$m and $\beta=1.8$. All of the sources fit in this way only had 2 undetected points, usually the SPIRE PLW and PMW bands.

```python
#Import modules and set plotting to be inside the notebook
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from glob import glob
import pandas as pd
%matplotlib inline
sns.set()
```

```python
#Make sure the path to the modeling modules is in the PythonPath
import sys
sys.path.append('/ricci9nb/tshimizu/Github/bat-agn-sed-fitting/')
```

```python
# Get all of the pickle files in the directory
pickle_files = glob('./pickles/*.pickle')
pickle_files
```

```python
# Loop over all of the files and pull out the best-fit parameters for each source
fit_data = pd.DataFrame(columns=['logMdust', 'Tdust', 'beta', 'logNpl', 'alpha', 'wturn'])
for f in pickle_files:
    name = f.split('/')[-1].split('_')[0]
    src_file = open(f, 'rb')
    data = pickle.load(src_file)
    fit_data.loc[name] = data['best_fit_model'].parameters
```

```python
# We can take a look at the DataFrame to make sure everything looks ok.
fit_data
```

```python
# Let's plot the pairwise relationship between all of the best fit parameters to see if there are any correlations
# between them.
fig = sns.PairGrid(fit_data, vars=['logMdust', 'Tdust', 'logNpl', 'alpha'])
fig.map_upper(plt.scatter)
fig.map_lower(sns.kdeplot)
fig.map_diag(sns.kdeplot, legend=False)
```

With only 42 sources its hard to see if there are any patterns in the best fit parameters. $T_{dust}$ and $M_{dust}$ are definitely still negatively correlated pretty strongly. Other than that it doesn't look like any others are that strongly correlated. The $\alpha$ parameter looks like it got pretty broad almost showing two peaks around 1.2 and 1.6. $T_{dust}$ also looks very broad with a big tail towards higher temperatures.$M_{dust}$ correspondingly has a pretty long tail towards lower dust masses.

```python
fit_data.describe(percentiles=[0.025, 0.16, 0.50, 0.84, 0.975])
```

Based on the statistics from this fitting run, things seem to have shifted a little bit. Median $T_{dust}$ went down by about 1 K so its not that big of a shift. Median $M_{dust}$ decreased by about 0.2 dex from 7.0 to 6.8 and $\alpha$ decreased from 1.6 to 1.48. Overall though they are still quite close to median values found from the other fitting runs. Some of the discrepancies are probably due to some of the bad fits that occurred in this run.
